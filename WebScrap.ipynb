```python
# function for reading a webpage

def read (urls):
    page = url(urls)
    pageHTML = page.read()
    page.close()
    ps = soup(pageHTML , 'html.parser')
    return ps
```

```python
# convert to integer
def conv (a):
    return int(''.join(a.split(',')))
```

```python
def Web(weburl):
    from IPython.display import clear_output
    page_main = read(weburl)     
    #finding the container that contains the desired data that we need
    df = pd.DataFrame()
    index=0
    all_data = page_main.find_all('a', {'class':'_31qSD5'})
    length = len(all_data)
    for data in all_data:
        dict_={}
        d=dict_
        d['Company'] = data.find('div',{'class':'_3wU53n'}).text.split()[0]
        #d['name'] = data.div.text

        #converting Data to integer data_type
        rating = data.find('span',{'class':'_38sUEc'}).text.split()
        d['NosRatings']=conv(rating[0])
        d['NosReviews']=conv(rating[3])


        total_rating = data.find('div',{'class':'hGSR34'}).text
        d["Rating"]=float(total_rating)

        d['CurrentCost']=conv(data.find('div',{'class':'_1vC4OE _2rQ-NK'}).text[1:])
        #d['ActualCost']=conv(data.find('div',{'class':'_3auQ3N _2GcJzG'}).text[1:])

        # Crawling into the individual sections
        # Collecting specific detials of each section

        detail_page_url = 'https://www.flipkart.com'+data['href']
        detail = read (detail_page_url)
        page_tables = detail.find('table',{'class':'_3ENrHu'})
        key = page_tables.tbody.find_all('td',{'class':'_3-wDH3 col col-3-12'})
        values = page_tables.tbody.find_all('li',{'class':'_3YhLQA'})
        for i in range (len(key)):
            try:
                try_val = conv(values[i].text)
                d[key[i].text]=try_val
            except:
                d[key[i].text]=values[i].text
        print (dict_)
        
        clear_output(wait=True)
        pop= ("*"*(index))+('..'*(length-index-1))
        print(pop)

        df1 = pd.DataFrame(dict_,index=[index])
        df = pd.concat([df,df1])
        index=index+1

        #storing it into data frame

    pd.set_option('display.max_columns', None)
    return (df)
```
